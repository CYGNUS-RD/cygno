```python
import pandas as pd
import numpy as np
def panda_from_gspreadsheet(key, sheet_name='Sheet1', header=0):
  # google spredsheet TO panda dataframe
  # ricordarsi di abilitare lo sharing pubblico dei dati

  url = 'https://docs.google.com/spreadsheets/d/{key}/gviz/tq?tqx=out:csv&sheet={sheet_name}&headers=1'.format(
      key=key, sheet_name=sheet_name.replace(' ', '%20'))
  return pd.read_csv(url, header=header)
dc =  panda_from_gspreadsheet('1eEDZST-gUF1IELlRzGcLxuxVkPNnKTLbI-eTdULwKU0', sheet_name='Sheet1')
dc = dc.sort_values(by=['Last Name'])
df = (dc[dc.skip != '#']).copy(deep=False) # skippa i commenti
df = df.reset_index()
df['name_initial'] = df['First Name'].str[0]
df['name_initial'] = df['name_initial'].apply(lambda x: "{}{}".format(x, '.'))
df['middle_initial'] = df['Middle Name']
for i, middle in enumerate(df['middle_initial'].values):
    names = ''
    if str(middle)!='nan':
        ms = middle.split(' ')
        for j, name in enumerate(ms):
            names = names+name[0]+'. '
    df.at[i, "middle_initial"]=names
authors = df[['name_initial', 'middle_initial', 'Last Name']].agg(' '.join, axis=1).values
```

```python
# print plan text format
', '.join(authors)
```

```python
# example of latex format
print('\\author{')
for i, author in enumerate(authors):
    print('\\firstname{'+'{0:s}{1:s}'.format(df['name_initial'].values[i], df['middle_initial'].values[i])+'\\lastname{'+'{0:s}'.format(df['Last Name'].values[i])+'}')
print('}')
```
